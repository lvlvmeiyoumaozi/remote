# Generating Data

subplots() can generate one or more plots in the same figure. <br>The variable *fig* represents the entire figure or collection of plots that are generated.<br>The variable *ax* represents a single plot in the figure and is the variable we'll use most of the time.

```python
import matplotlib.pyplot as plt
squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth = 3)
ax.set_title("square numbers", fontsize = 24)
ax.set_xlabel("value", fontsize = 14)
ax.set_ylabel("square of value", fontsize = 14)
ax.tick_params(axis='both', labelsize = 14)
plt.show()
```

```python
x_values = range(1,1001)
y_values = [x**2 for x in x_values]
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=(0,0.8,0), s=10)
ax.axis([0,1100,0,1100000])
plt.show()
```

```python
x_values = range(1,1001)
y_values = [x**2 for x in x_values]
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, s=10)
ax.axis([0,1100,0,1100000])
plt.show()
```

**cmap**: This code colors the points with lower y-values light blue and colors the points with higher y-values dark blue.<br>the name of color must add *s*

```python
x_values = range(1,5000)
y_values = [x**3 for x in x_values]
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Greens, s=10)
plt.savefig('saquares_plot1.png')
```
